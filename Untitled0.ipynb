```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sb 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv("/content/drive/MyDrive/anshika didi mtech/heart.csv")
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.iloc[:,:-1],data["target"],test_size=.2,random_state=42)  # training data 80 % and testing data 20%
```

KNN

```python
model_knn=KNeighborsClassifier(n_neighbors=3)
model_knn.fit(x_train,y_train)
pred_knn=model_knn.predict(x_test)
print("accuracy score :-  ",accuracy_score(y_test,pred_knn))
print("confusion matrix")
print(confusion_matrix(y_test,pred_knn))
print("Classification report")
print(classification_report(y_test,pred_knn))
```

SVM

```python
model_svc=SVC()
model_svc.fit(x_train,y_train)
pred_svc=model_svc.predict(x_test)
print("accuracy score :-  ",accuracy_score(y_test,pred_svc))
print("confusion matrix")
print(confusion_matrix(y_test,pred_svc))
print("Classification report")
print(classification_report(y_test,pred_svc))
```


naive_bayes

```python
model_mnb=MultinomialNB()
model_mnb.fit(x_train,y_train)
pred_mnb=model_mnb.predict(x_test)
print("accuracy score :-  ",accuracy_score(y_test,pred_mnb))
print("confusion matrix")
print(confusion_matrix(y_test,pred_mnb))
print("Classification report")
print(classification_report(y_test,pred_mnb))
```

DecisionTree

```python
model_dcs=DecisionTreeClassifier(max_depth=500)
model_dcs.fit(x_train,y_train)
pred_dcs=model_dcs.predict(x_test)
print("accuracy score :-  ",accuracy_score(y_test,pred_dcs))
print("confusion matrix")
print(confusion_matrix(y_test,pred_dcs))
print("Classification report")
print(classification_report(y_test,pred_dcs))
```
